```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('students.csv')

# Look at the data
print(df.head())
print(df.info())
```

```python
# Handling missing values
print(df.isnull().sum())
df['Grade'].fillna('Unknown', inplace=True)
df.dropna(subset=['Age'], inplace=True)
```

```python
# Fixing data types and cleaning text
df['Age'] = df['Age'].astype(int)
df['Name'] = df['Name'].str.strip().str.title()
```

```python
# Removing duplicates and creating new column
df.drop_duplicates(inplace=True)
df['Passed'] = df['Grade'].isin(['A', 'B', 'C'])
print(df[['Name', 'Grade', 'Passed']])
```

```python
# Histogram of Age
df['Age'].plot(kind='hist', bins=5, title='Age Distribution')
```
